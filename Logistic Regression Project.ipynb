```python
crash_data = pd.read_csv('Traffic_Crashes_-_Crashes.csv')
new_cols_data = crash_data[["TRAFFICWAY_TYPE", "ROADWAY_SURFACE_COND", "LIGHTING_CONDITION", "POSTED_SPEED_LIMIT", "ROAD_DEFECT", "MOST_SEVERE_INJURY", "INJURIES_FATAL"]].copy()
new_cols_data_df = new_cols_data.dropna()
new_cols_data_df["INJURIES_FATAL"].replace(0, 'non-fatal', inplace=True)
new_cols_data_df["INJURIES_FATAL"].replace(1, 'fatal', inplace=True)
new_cols_data_df["INJURIES_FATAL"].replace(2, 'fatal', inplace=True)
new_cols_data_df["INJURIES_FATAL"].replace(3, 'fatal', inplace=True)
count = new_cols_data_df["INJURIES_FATAL"].unique()
count
new_cols_data_df
```

```python
index_names = new_cols_data_df[ (new_cols_data_df['MOST_SEVERE_INJURY'] == 'NO INDICATION OF INJURY') ].index
new_cols_data_df.drop(index_names , inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
new_df = new_cols_data_df.apply(LabelEncoder().fit_transform)
new_df
```

```python
from keras.utils import to_categorical
```

```python
new_hot_df = to_categorical(new_df)
new_hot_df
```

```python
X = new_df.drop(["INJURIES_FATAL"], axis=1)
y = new_cols_data_df["INJURIES_FATAL"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```
